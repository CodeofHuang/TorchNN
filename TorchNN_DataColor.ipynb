```python
import os
import cv2
import matplotlib.pyplot as plt
import torchvision.transforms as transforms
from PIL import Image

def random_color_jitter(pil_image, brightness=0.1, contrast=0.1,
                        saturation=0.1, hue=0.1, isShow=True):
  transform=transforms.Compose([
      transforms.ColorJitter(brightness=brightness,contrast=contrast,saturation=saturation,hue=hue)
      ])
  color_jitter_pil = transform(pil_image)
  if isShow:
    plt.imshow(color_jitter_pil)
    plt.show()
  return color_jitter_pil

img_path = "/content/drive/My Drive/demoDATA/testdemoOSCD.png"
pil_img = cv2.imread(img_path)
pil_img = Image.fromarray(pil_img)
b,c,s,h=0,0,0,0.05
for i in range(3):
  random_color_jitter(pil_img, brightness=b, contrast=c, saturation=s, hue=h, isShow=True)
```
