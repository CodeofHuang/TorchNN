```python
## Batch Normalization
import torch.nn as nn
import torch
m = nn.BatchNorm2d(3) #channel数量
input = torch.randn(1,3,5,5) #Normal(0,1)
output = m(input)
print(input)
print(output)
input = torch.rand(1,3,5,5) #Rand Uniform(0,1)
output = m(input)
print(input)
print(output)
```
